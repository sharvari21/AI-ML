```python
data = pd.read_csv('BreadBasket_DMS - Arindam Dev.csv')
data.head()
```

```python
data['Quantity'] = np.nan
data['Quantity'] = data['Quantity'].fillna(1)
```

```python
data.head()
```

```python
data = data[data['Item'] != 'NONE']
```

```python
data
```

```python
data1 = data[['Transaction','Item','Quantity']]
data1.head()
```

```python
data2 = data1.groupby(['Transaction','Item'])['Quantity'].sum()
data2
```

```python
data3 = data2.unstack()
```

```python
data4 = data3.applymap(lambda x : 1 if x>0 else 0)
data4.head()
```

```python
support = apriori(data4, min_support= 0.02, use_colnames=True)
support
```

```python
association = association_rules(support, metric='lift', min_threshold=1)
```

```python
association.sort_values(by = 'lift', ascending = False)
```
